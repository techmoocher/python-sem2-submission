#Bài tập 2: Thực hành với mảng 2 chiều

Kết quả của một kỳ thi tuyển sinh được công bố có cấu trúc: kết quả mỗi thí sinh trên 1 dòng gồm 4 số thực là điểm các môn Toán, Văn, Anh và môn chuyên. Các thí sinh được đánh thứ tự từ 1.

Điều kiện trúng tuyển như sau:
- Không có môn nào dưới 2.5.
- Điểm xét tuyển được tính là tổng điểm thi các môn với môn chuyên có hệ số 2.
- Điểm xét tuyển phải từ 25 trở lên.

Đọc dữ liệu từ file văn bản `input.txt` và lưu thành một mảng 2 chiều `candidates`

```python
def loadFile(filename):
    finp = open(filename, "r")
    lines = finp.readlines()
    arr2D = []
    for line in lines:
        arr2D.append(list(map(float, line.split())))
    finp.close()
    return arr2D

candidates = loadFile("/content/sample_data/input.txt")
print(candidates)
```

##a. In ra danh sách điểm trúng tuyển, mỗi điểm trên một dòng. Danh sách có thứ tự giảm dần theo điểm trúng tuyển.

```python
def admittedList(candidates):
    #start your code here
    admitted = []
    for candidate in candidates:
        total = 0
        if min(candidate) < 2.5:
            continue
        for i in range(3):
            total += candidate[i]
        total += candidate[3] * 2
        if total >= 25.0:
            admitted.append(total)

    return admitted
    #end your code here

passed = admittedList(candidates)
for v in sorted(passed, reverse=True):
    print(v)
```

##b. Tính tỉ lệ % trúng tuyển trên tổng số thí sinh dự thi. Kết quả làm tròn 2 chữ số thập phân.

```python
def passRate(candidates):
    #start your code here
    admitted = 0
    for candidate in candidates:
        total = 0
        if min(candidate) < 2.5:
            continue
        for i in range(3):
            total += candidate[i]
        total += candidate[3] * 2
        if total >= 25.0:
            admitted += 1
    return admitted / len(candidates)
    #end your code here

print(format(passRate(candidates)*100, ".2f"))
```

##c. Tìm danh sách thứ tự thí sinh không trúng tuyển và có điểm môn chuyên >= 7

```python
def failedCandidates(candidates):
    #start your code here
    unlucky = []
    for i in range(len(candidates)):
        total = 0
        for j in range(3):
            total += candidates[i][j]
        if min(candidates[i][:4]) >= 2.5 and total + 2 * candidates[i][3] >= 25:
            continue
        if candidates[i][3] >= 7:
            unlucky.append(i+1)
    return unlucky
    #end your code here

for p in failedCandidates(candidates):
    print(p)
```

##d. Tìm danh sách thứ tự thí sinh có điểm tổng >= 25 nhưng bị ít nhất 1 môn thi có điểm < 2.5

```python
def failGrade(candidates):
    #start your code here
    unfortunate = []
    for i in range(len(candidates)):
        total = 0
        for j in range(3):
            total += candidates[i][j]
        if total + candidates[i][3] * 2 >= 25:
            for score in candidates[i]:
                if score < 2.5:
                    unfortunate.append(i+1)
                    break
    return unfortunate
    #end your code here

for p in failGrade(candidates):
    print(p)
```
